```python
import pandas as pd

# Assuming your CSV file is named 'data.csv' and is in the same directory as your notebook
# If not, replace 'data.csv' with the correct path to your file.
try:
    df = pd.read_csv('/content/EMP_TEMP_SEX_ECO_NB_A-20251107T0859.csv', on_bad_lines='skip', quotechar='"', doublequote=True)

    # Display details of the DataFrame
    print("DataFrame Info:")
    display(df.info())

    # Display the top 5 rows of the DataFrame
    print("\nTop 5 Rows:")
    display(df.head())

except FileNotFoundError:
    print("Error: 'data.csv' not found. Please make sure the file is in the correct directory.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Filter for 'sex' equal to 'SEX_T'
df_filtered = df[df['sex'] == 'SEX_T'].copy()

# Filter for 'classif1' starting with 'ECO_ISIC4_'
df_filtered = df_filtered[df_filtered['classif1'].str.startswith('ECO_ISIC4_')]

# Drop specified columns
columns_to_drop = ['obs_status', 'note_classif', 'note_indicator', 'note_source', 'source', 'sex']
df_filtered = df_filtered.drop(columns=columns_to_drop)

# Display the filtered DataFrame
print("\nFiltered DataFrame:")
display(df_filtered.head())
```

```python
# Group by 'time', 'indicator', and 'classif1' and sum 'obs_value'
df_grouped = df_filtered.groupby(['time', 'indicator', 'classif1'])['obs_value'].sum().reset_index()

# Display the grouped DataFrame
print("\nGrouped DataFrame (Sum of obs_value across ref_area):")
display(df_grouped.head())
```

```python
# Sort by 'classif1' and 'time' to ensure correct growth rate calculation
df_grouped = df_grouped.sort_values(by=['classif1', 'time'])

# Calculate the year-over-year growth rate for each classif1
df_grouped['growth_rate'] = df_grouped.groupby('classif1')['obs_value'].pct_change()

# Remove rows before 2023
df_grouped = df_grouped[df_grouped['time'] == 2022]

# Reset the index
df_grouped = df_grouped.reset_index(drop=True)

# Rename columns
df_grouped = df_grouped.rename(columns={
    'indicator': 'indicator_type_id',
    'classif1': 'industry_id',
    'time': 'last_updated',
    'growth_rate': 'value'
})

# Display the updated DataFrame
print("\nDataFrame with Growth Rate and Renamed Columns:")
display(df_grouped.head(50))
```
